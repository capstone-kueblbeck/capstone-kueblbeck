```python
import pandas as pd

def get_location_data(location, year):
    columns = ['lfnr', 'lieferant', 'artnr', 'beschreibung', 'index',
               'ltz_vk_ges', 'basispreis', 'basispr_summe', 'gesamt_lager']
    
    location_columns = columns + [f'{location.lower()}_lager',
                                   f'ltz_vk_{location.lower()}',
                                   'gesamt_vk',
                                   f'{location.lower()}_vk']
    
    filtered_data = df_master[location_columns]
    filtered_data = filtered_data[(filtered_data['gesamt_lager'] == 0) & (filtered_data[f'{location.lower()}_lager'] == 0)]
    
    # Convert the ltz_vk column to datetime
    filtered_data[f'ltz_vk_{location.lower()}'] = pd.to_datetime(filtered_data[f'ltz_vk_{location.lower()}'], errors='coerce')
    
    filtered_data = filtered_data[filtered_data[f'ltz_vk_{location.lower()}'].dt.year == year]
    
    return filtered_data


location_data = get_location_data('WEN', 2002)
location_data


```
